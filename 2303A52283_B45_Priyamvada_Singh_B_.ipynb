**2.Data Visualization: Plot feature
distributions for pH, alcohol, and citric
acid.**

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')


fig, axes = plt.subplots(1, 3, figsize=(15, 5))


sns.histplot(data['pH'], kde=True, ax=axes[0])
axes[0].set_title('pH Distribution')


sns.histplot(data['alcohol'], kde=True, ax=axes[1])
axes[1].set_title('Alcohol Distribution')


sns.histplot(data['citric acid'], kde=True, ax=axes[2])
axes[2].set_title('Citric Acid Distribution')


plt.tight_layout()
plt.show()
```
